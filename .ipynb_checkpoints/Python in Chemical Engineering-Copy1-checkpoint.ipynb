# Flow through orifice

```python
import numpy as np
from matplotlib import pyplot as plt
from math import pi

#input data
d = float(input("Diameter of Orifice (mm): "))
h = float(input("Height of tank (mm): "))

#area calculation
a = (pi/4)*d**2 #mm2
print ("Area of orifice:",a,"mm2")

#mass flowrate calculation
x = np.linspace(0,h,20)
Q = (0.61*a)*((2*9810*x)**(1/2))

#for plotting results
plt.xlabel("Q volume flowrate (m3/s)")
plt.ylabel("Height (m)")
plt.plot(Q,x,label="Flow over time ")
plt.legend()
```
